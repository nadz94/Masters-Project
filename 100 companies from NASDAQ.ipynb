```python
monthly_data = Real_stock_analysis(monthly_csv)
weekly_data = Real_stock_analysis(weekly_csv)
daily_data = Real_stock_analysis(daily_csv)
```

```python
jb = [monthly_data.jb_100(),
weekly_data.jb_100(),
daily_data.jb_100()]
```

```python
plt.boxplot(jb, labels = ['monthly', 'weekly', 'daily'])
plt.legend()
plt.ylim(0,6000)
plt.title('Jarque-Bera values for returns of 100 NASDAQ companies')
plt.ylabel('JB Test Value')
plt.xlabel('Frequency')
plt.show()
```

```python
plt.boxplot(jb, labels = ['monthly', 'weekly', 'daily'])
plt.plot(range(5),[5.99]*5, c='green', label = 'JB critical value')
plt.legend()
plt.ylim(0,40)
plt.title('Jarque-Bera values for returns of 100 NASDAQ companies')
plt.ylabel('JB Test Value')
plt.xlabel('Frequency')
plt.show()
```

```python
auto_correlation = [monthly_data.standard_autocorr_values(monthly_data.returns()), \
        weekly_data.standard_autocorr_values(weekly_data.returns()), daily_data.standard_autocorr_values(daily_data.returns())]
```

```python
plt.boxplot(auto_correlation, labels = ['monthly', 'weekly', 'daily'])
m = 1.96 / math.sqrt(12-1)
w = 1.96 / math.sqrt(53-1)
d = 1.96 / math.sqrt(252-1)
plt.plot(range(5),[m]*5, c='green', label = 'Monthly Critical Value')
plt.plot(range(5),[w]*5, c='orange', label = 'Weekly Critical Value')
plt.plot(range(5),[d]*5, c='magenta', label = 'Daily Critical Value')
plt.legend()
plt.title('Auto-correlation of Returns')
plt.ylabel('Correlation value')
plt.xlabel('Frequency')
plt.show()
```

```python
abs_auto_correlation = [monthly_data.abs_autocorr_values(monthly_data.returns()), \
        weekly_data.abs_autocorr_values(weekly_data.returns()), daily_data.abs_autocorr_values(daily_data.returns())]
```
