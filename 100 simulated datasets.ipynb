```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#box plot for jarque-bera test

def jb_boxplot(monthly, weekly, daily):
    jb = [monthly.jb_100(), weekly.jb_100(), daily.jb_100()]
    plt.boxplot(jb, labels = ['monthly', 'weekly', 'daily'])
    plt.title('Jarque-Bera Values')
    plt.plot(range(5),[5.99]*5, c='green', label = 'JB critical value')
    plt.ylabel('JB Test Value')
    plt.xlabel('Frequency')
    plt.legend()
    plt.ylim(-1,10)
    plt.show()
```

```python
def autocorr_boxplot(monthly, weekly, daily):
    autocorr = [monthly.standard_autocorr_values(), weekly.standard_autocorr_values(), daily.standard_autocorr_values()]
    plt.boxplot(autocorr, labels = ['monthly', 'weekly', 'daily'])
    plt.title('Auto-correlation of Returns')
    plt.ylabel('Correlation value')
    plt.xlabel('Frequency')
    plt.show()
```

```python
def sqrd_autcorr_boxplot(monthly, weekly, daily):
    sqrd_autocorr = [monthly.sqrd_autocorr_values(), weekly.sqrd_autocorr_values(), daily.sqrd_autocorr_values()]
    plt.boxplot(sqrd_autocorr, labels = ['monthly', 'weekly', 'daily'])
    plt.title('Auto-correlation of Squared Returns')
    plt.ylabel('Correlation value')
    plt.xlabel('Frequency')
    plt.show()
```
